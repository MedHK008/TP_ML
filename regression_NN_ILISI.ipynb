# TP6:réseau de neurone pour la régression


```python
import numpy as np
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python

# LOAD DATA
dataset = loadtxt('pima-indians-diabetes.csv',skiprows=1,delimiter=',')
x = dataset[:,0:8]
y = dataset[:,8]
```

```python
model = Sequential()
model.add(Dense(12, input_shape=(8,), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x, y, epochs=150, batch_size=10)
```

```python
_,accuracy = model.evaluate(x, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
Xnew = np.array([[1,103,30,38,83,43.3,0.183,33]])

prediction = model.predict(Xnew)
rounded = [round(x[0]) for x in prediction]
print("prediction for 1,103,30,38,83,43.3,0.183,33 is: ",rounded)
```

```python
# save the model
model.save('model.keras')
```
