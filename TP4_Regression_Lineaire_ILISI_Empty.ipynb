1- Génération automatique de données pour l'entraînement et le test

a) Pour générer les données aléatoiremment (100 observations) suivant la relation linéaire: Y = 2 + 3 * X utiliser le code suivant:

import numpy as np

np.random.seed(42)

X = 2 * np.random.rand(100, 1)  # valeurs générées selon la loi uniforme entre 0 et 2

Y = 2 - 3 * X + np.random.randn(100, 1)  # Relation linéaire avec bruit généré selon la loi uniforme

b) afficher le type et la valeur de X et Y


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

np.random.seed(42)

X = 2 * np.random.rand(100, 1) 

Y = 2 - 3 * X + np.random.randn(100, 1)

print("Type de X:", type(X))
print("Valeur de X:\n", X)
print("Type de Y:", type(Y))
print("Valeur de Y:\n", Y)
```

2- Visualisation de données

En utilisant le type de graphique "scatter" de "Matplotlib" afficher le nuage de points correspondant aux données générées


```python

plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter plot of generated data')
plt.show()
```

3- diviser l'ensemble de données (X, Y) en deux ensembles: données d'entrainement (X_train, X_train) et données de test (X_test, Y_test) en utilisant split_ratio = 0.8

```python
#  Ajouter le code pour Séparer en train et test ...
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42,shuffle=True)
```

4- Implémentation de la Régression linéaire analytique (Moindres carrés ordinaires)

Implémenter la méthode "analytical_solution" des Moindres carrés en utilisant la formule 

```python
# Ajouter le code pour l'implémentation de "analytical_solution" ...
def analytical_solution(X, y):
    x_b = np.c_[np.ones((len(X), 1)), X]
    theta_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)
    return theta_best

theta = analytical_solution(x_train, y_train)
print("Theta_best:", theta)
```

5- implémenter une fonction pour prédire les sorties Y_pred_analytical des entrées X_test en utilisant le modèle obtenu 

```python
# Ajouter le code pour la fonction de Prédictions ...
def predict(X, theta):
    x_b = np.c_[np.ones((len(X), 1)), X]
    y_predict = x_b.dot(theta)
    return y_predict

y_pred_analytical = predict(x_test, theta)
print("Y_pred_analytical:", y_pred_analytical)
```

6- Evaluer le modèle obtenu en utilsant les métriques MSE et R2

utiliser les fonctions mean_squared_error et r2_score importé de sklearn.metrics

```python

# Ajouter le code pour l'évaluation du modèle ...
mse = mean_squared_error(y_test, y_pred_analytical)
r2 = r2_score(y_test, y_pred_analytical)

print("Mean Squared Error (MSE):", mse)
print("Coefficient of Determination (R2):", r2)
```

7- Afficher le modèle obtenu avec les données d'entrainement


```python
# Visualisation du modèle ...


plt.scatter(x_train, y_train, color='blue', label='Training data')
plt.plot(x_test, y_pred_analytical, color='red', label='Predictions')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Analytical Solution')
plt.legend()
plt.show()
```

8- Afficher le modèle obtenu avec les données de test

```python
# Visualisation du modèle ...
plt.scatter(x_test, y_test, color='green', label='Testing data')
plt.plot(x_test, y_pred_analytical, color='red', label='Predictions')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Analytical Solution')
plt.legend()
plt.show()
```

II- implémentation de la régression linéaire en utilisant la méthode de la descente de gradient

Implémenter une fonction "gradient_descent" pour estimer les coefficients (beta) du modèle en utilisant les paramètres suivants:

lr = 0.1 (taux d'apprentissage)

n_iterations=1000 (nombre d'itérations)

```python
#  Ajouter le code pour la méthode de la Descente de gradient ...

def gradient_descent(X, y, lr=0.1, n_iterations=1000):
    m = len(y)
    theta = np.random.randn(2,1)
    X_b = np.c_[np.ones((m, 1)), X] 
    
    for iteration in range(n_iterations):
        gradients = 2/m * X_b.T @ (X_b @ theta - y)
        theta = theta - lr * gradients
        
    return theta

theta_gd = gradient_descent(x_train, y_train)
print("Theta_gd:", theta_gd)

```

```python
# Prédictions ...
def predict(X, theta):
    X_b = np.c_[np.ones((X.shape[0], 1)), X]  # add x0 = 1 to each instance
    return X_b @ theta

y_pred_gd = predict(x_test, theta_gd)
print("Y_pred_gd:", y_pred_gd)
```

```python
# évaluation du modèle ...
mse = mean_squared_error(y_test, y_pred_gd)

r2 = r2_score(y_test, y_pred_gd)

print("Mean Squared Error (MSE):", mse)

print("Coefficient of Determination (R2):", r2)
```

```python
# Visualisation du modèle ...
plt.scatter(x_train, y_train, color='blue', label='Training data')

plt.plot(x_train, predict(x_train, theta_gd), color='red', label='Fitted line')

plt.xlabel('X')

plt.ylabel('Y')

plt.title('Linear Regression Model - Training Data')

plt.legend()

plt.show()
```

```python
# Visualisation du modèle ...
plt.scatter(x_test, y_test, color='blue', label='Training data')
plt.plot(x_test, predict(x_test, theta_gd), color='red', label='Fitted line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression Model - Training Data')
plt.legend()
plt.show()
```
