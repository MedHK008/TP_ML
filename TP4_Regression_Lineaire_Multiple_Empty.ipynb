```python
# Charger les données
data = pd.read_csv('donnees_immobilieres.csv')

# Afficher les premières lignes

# Statistiques descriptives

#Vérifier les corrélations entre les variables
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matrice de Corrélation')
plt.show()
```

3- Séparer les variables explicatives (X) et la variable cible (y) en utilisant "train_test_split" de "sklearn"
- Diviser les données en ensembles d'entraînement et de test (80 % / 20 %) 
- Afficher la taille de l'ensemble d'entraînement et de l'ensemble de test

```python
# Corriger les noms de colonnes
data.columns = data.columns.str.strip()

# Séparer les variables explicatives (X) et la variable cible (y)
x = data[['Surface', 'Chambres','Age']]
y = data['Prix']
# Diviser les données en ensembles d'entraînement et de test (80 % / 20 %)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, shuffle=True, stratify=None)     
# Afficher la taille de l'ensemble d'entraînement et de l'ensemble de test
print(f"Taille de l'ensemble d'entraînement: {x_train.shape[0]} échantillons")
print(f"Taille de l'ensemble de test: {x_test.shape[0]} échantillons")
```

4-  Initialiser et entraîner le modèle en utilisant  "LinearRegression" de "sklearn"

```python
# Initialiser et entraîner le modèle
model = LinearRegression()
model.fit(x_train, y_train)
# Coefficients du modèle
print(f'Coefficients: {model.coef_}')
```
