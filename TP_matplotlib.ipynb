# TP: MODULE MATPLOTLIB

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ventes.csv')

df.head()

df.info()
df.describe()
```

## 2. Visualisations temporelles (Graphiques de courbe)

```python
sofia_sales = df[df['Vendeur'] == 'Sofia'].groupby('Mois')['Ventes'].sum()
plt.plot(sofia_sales.index, sofia_sales.values)
plt.xlabel('Mois')
plt.ylabel('Ventes')
plt.show()
```

```python
monthly_sales = df.groupby('Mois')['Ventes'].mean()
plt.plot(monthly_sales.index, monthly_sales.values)
plt.xlabel('Mois')
plt.ylabel('Ventes moyennes')
plt.show()
```

```python
casablanca_sales = df[df['Ville'] == 'Casablanca'].groupby('Mois')['Ventes'].mean()
rabat_sales = df[df['Ville'] == 'Rabat'].groupby('Mois')['Ventes'].mean()
plt.plot(casablanca_sales.index, casablanca_sales.values, label='Casablanca')
plt.plot(rabat_sales.index, rabat_sales.values, label='Rabat')
plt.xlabel('Mois')
plt.ylabel('Ventes moyennes')
plt.legend()
plt.show()
```

## 3. Comparaisons entre vendeurs (Histogrammes)

```python
annual_sales_by_seller = df.groupby('Vendeur')['Ventes'].sum()
annual_sales_by_seller.plot(kind='bar')
plt.xlabel('Vendeur')
plt.ylabel('Ventes totales')
plt.show()
```

```python
annual_sales_by_city = df.groupby('Ville')['Ventes'].sum()
annual_sales_by_city.plot(kind='bar')
plt.xlabel('Ville')
plt.ylabel('Ventes totales')
plt.show()
```

```python
df.groupby('Vendeur')['Ventes'].sum().plot(kind='bar', alpha=0.5, label='Vendeur')
plt.xlabel('Vendeur')
plt.ylabel('Ventes totales')
plt.show()
```

## 4. Rpartition des donnes (Graphiques en secteurs)

```python
annual_sales_by_city = df.groupby('Ville')['Ventes'].sum()
annual_sales_by_city.plot(kind='pie', autopct='%1.1f%%')
plt.title('Part du total des ventes annuelles pour chaque ville')
plt.show()
```

```python
march_sales = df[df['Mois'] == 3].groupby('Vendeur')['Ventes'].sum()
march_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Comparaison des ventes mensuelles en Mars entre les vendeurs')
plt.show()
```
